### **Exercise1:**

Give students in the `example.csv` file a bonus of 5 for all the courses. Find a way to replace all the values that are now greater than 30 by 30. Save the new data to a file. Make a plot with histograms for every course in the example file.

```python
import pandas as pd
import matplotlib.pyplot as plt

# read the csv file
df = pd.read_csv("example.csv")

#print(df)
# add 5 to all values in the score column
for i in range(1,5):
  df['course'+str(i)] += 5
#print(df)

# replace all values greater than 30 with 30
for i in range(1,5):
  df['course'+str(i)] = df['course'+str(i)].clip(upper=30)
#print(df)

# save the new data to a new file
df.to_csv("new_example.csv", index=False)

# make a histogram for each course
for i in range(1,5):
    df['course'+str(i)].plot(kind='hist', title='histo')
    plt.show()
```

### **Exercise3**
 - create a random matrix with numpy
 - save it to file
 - read it back with numpy
 - calculate some statistcs on it with numpy 
 - read the same file with pandas
 - calculate the same statistics on it with pandas

```python
import numpy as np
import pandas as pd

# create a random matrix with numpy
matrix = np.random.rand(50,1)

# save it to file
np.savetxt("matrix.csv", matrix)

# read it back with numpy
loaded_matrix = np.loadtxt("matrix.csv")

# calculate some statistics on it with numpy
print("Mean:", np.mean(loaded_matrix))
print("Standard deviation:", np.std(loaded_matrix))

# read the same file with pandas
df = pd.read_csv("matrix.csv", header=None)

# calculate the same statistics on it with pandas
print("Mean:", df.mean(numeric_only=True))
print("Standard deviation:", df.std(numeric_only=True))
```
